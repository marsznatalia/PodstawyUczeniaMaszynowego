```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/content/PodstawyUczeniaMaszynowego/Salary Data (2).csv')

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Years of Experience', y='Salary', data=data)
plt.title('Years of Experience vs. Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
average_salary_by_experience = data.groupby('Years of Experience')['Salary'].mean()

plt.figure(figsize=(8, 6))
plt.plot(average_salary_by_experience.index, average_salary_by_experience.values)
plt.title('Average Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary')
plt.grid(True)
plt.show()
```
